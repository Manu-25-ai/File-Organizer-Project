#  File Organizer – A script that sorts files into folders by file type.

```python
import os
import shutil
```

```python
folder_path=input("Enter you folder path:")
```

```python
file_types = {
    "Images": [".jpg", ".jpeg", ".png", ".gif"],
    "Documents": [".pdf", ".docx", ".txt", ".pptx", ".xls", ".xlsx"],
    "Videos": [".mp4", ".mkv", ".mov", ".avi"],
    "Music": [".mp3", ".wav", ".aac"],
    "Archives": [".zip", ".rar", ".7z"],
    "Programs": [".exe", ".msi", ".py", ".java", ".c", ".cpp"]
}
```

```python
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    # Skip folders
    if os.path.isdir(file_path):
        continue

    # Get the file extension
    _, extension = os.path.splitext(filename)

    # Match file type
    for folder_name, extensions in file_types.items():
        if extension.lower() in extensions:
            folder_name_path = os.path.join(folder_path, folder_name)
            
            # Create folder if it doesn't exist
            if not os.path.exists(folder_name_path):
                os.makedirs(folder_name_path)
            
            # Move file
            shutil.move(file_path, folder_name_path)
            print(f"Moved: {filename} → {folder_name}/")
            break
print("File organized successfully")
```
